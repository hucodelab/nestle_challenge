# AdiÃ§Ã£o da DB tb_vendors

```python
df_vendors = pd.read_csv(
    'Nestle_VDB_Data_Challenge/tb_vendors.csv',
    sep=';')

df_vendors.avaliacao_media.unique()
df_vendors = df_vendors[['id_fornecedor', 'data_registro', 'delivery_raio_km', 'delivery_sistema', 
                         'hora_abertura', 'hora_fechamento']]
```

```python
df_vendors['data_registro'] = pd.to_datetime(df_vendors['data_registro'], format='%Y-%m-%d')
```

```python
df_vendors['today'] = pd.to_datetime("now")
```

```python
df_vendors['dias_site'] = df_vendors['today'] - df_vendors['data_registro']
```

```python
df_vendors['dias_site'] = df_vendors['dias_site'].dt.days
```

```python
def calc_tempo_aberto():
    min_abertura = []
    for minab in df_vendors['hora_abertura']:
        min_abertura.append((int(minab[:2])*60) + int(minab[3:]))

    min_fechamento = []
    for minfe in df_vendors['hora_fechamento']:
        min_fechamento.append((int(minfe[:2])*60) + int(minfe[3:]))
                              
    tempo_aberto = []
                              
    for i in range(len(min_abertura)):
        if min_abertura[i] < min_fechamento[i]:
            tempo_aberto.append((min_fechamento[i] - min_abertura[i])/60)
        else:
            tempo_aberto.append(1440 - (min_abertura[i] - min_fechamento[i])/60)
                              
    df_vendors['tempo_aberto_(h)'] = tempo_aberto
                              
    return df_vendors['tempo_aberto_(h)']        
```

```python
df_vendors['tempo_aberto_(h)'] = calc_tempo_aberto()
```

Foi feita a criaÃ§Ã£o de novas variÃ¡veis partindo de variÃ¡veis jÃ¡ existentes. Novas variÃ¡veis:

- dias_site: tempo total em dias de cada um dos fornecedores (que jÃ¡ venderam) no site desde a data de registro
- tempo_aberto_(h): tempo total de horas disponÃ­veis de cada fornecedor por dia

```python
df_vendors = df_vendors[['id_fornecedor', 'delivery_raio_km', 
                         'delivery_sistema', 'dias_site', 'tempo_aberto_(h)']]
```

# Tipo de endereÃ§o

```python
def plot_tipo_endereco():
    df = df_vendor_products_addresses.groupby(["tipo"]).mean()['subtotal']

    sns.set_theme(style="whitegrid")
    ax = sns.barplot(x=df.index, y=df, data=df)
    ax.set_xticklabels(['HISTORY', 'OFFICE'])
    ax.set_xlabel('')
    ax.set_ylabel('subtotal vendido (R$)')
    ax.set_title('subtotal vendido por tipo de endereÃ§o')
    
plot_tipo_endereco()
```

Podemos observar que o tipo de endereÃ§o Ã© uma variÃ¡vel importante, pois os fornecedores com tipo de endereÃ§o "OFFICE" tem vendido muito mais em R$ do que os fornecedores com tipo de endereÃ§o "HISTORY"

# Tipo da taxa de entrega

```python
def plot_tipo_taxa_entrega():
    df = df_vendor_products_addresses.groupby(["tipo_taxa"]).mean()['subtotal']

    sns.set_theme(style="whitegrid")
    ax = sns.barplot(x=df.index, y=df, data=df)
    ax.set_xticklabels(df.index)
    ax.set_xlabel('')
    ax.set_ylabel('subtotal vendido (R$)')
    ax.set_title('subtotal vendido por tipo de taxa de entrega')
    
plot_tipo_taxa_entrega()
```

Os fornecedores com o tipo de taxa de entrega "PER_KM" tem vendido mais que os fornecedores com outros tipos de taxa de entrega
